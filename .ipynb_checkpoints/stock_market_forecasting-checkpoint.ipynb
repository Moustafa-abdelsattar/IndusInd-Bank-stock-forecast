# Dataset and Feature Description

## Overview of the Dataset

#### Stock Market Data Description

Stock market data is widely analyzed for educational, business, and personal interests.
#### Link > https://www.kaggle.com/datasets/rohanrao/nifty50-stock-market-data
#### Content
The data is the price history and trading volumes of the fifty stocks in the index **NIFTY 50** from **NSE (National Stock Exchange)** India. All datasets are at a day-level with pricing and trading values split across `.csv` files for each stock along with a metadata file containing macro-information about the stocks themselves. 

The data spans from **1st January, 2000** to **30th April, 2021**.

#### I choosed INDUSINDBK.csv

## Feature Descriptions



- **Series**: Equity series of stock market (denoted as EQ).
- **Prev Close**: The closing price of the stock for the previous day.
- **Open**: The opening price of INDUSINDBK shares on the current day.
- **High**: The highest price of INDUSINDBK shares on the current day.
- **Low**: The lowest price of INDUSINDBK shares on the current day.
- **Last**: The last price of INDUSINDBK shares on the current day.
- **Close**: The closing price of INDUSINDBK shares on the current day.
- **VWAP**: Volume Weighted Average Price, the target variable to predict. It is a trading benchmark that provides the average price at which the stock has traded throughout the day, based on both volume and price.
- **Volume**: Volume of shares traded on the current day.
- **Turnover**: A measure of stock liquidity, calculated by dividing the total number of shares traded over a period by the average number of shares outstanding for the period.
- **Trades**: Total number of trades on the current day.
- **Deliverable Volume**: The quantity of shares that actually move from one set of people to another set of people.
- **Deliverable (%)**: Deliverable volume as a percentage.

## Stationarizing the Data

```python
# Apply Box-Cox transformation to the VWAP column
df['vwap_boxcox'], lam = boxcox(df['VWAP'])

# Print the lambda value for reference (it shows the optimal transformation parameter)
print(f"Optimal Lambda: {lam}")

# Plot the original VWAP vs the Box-Cox transformed data
plt.figure(figsize=(12, 6))

# Plot the original VWAP
plt.subplot(1, 2, 1)
plt.plot(df.index, df['VWAP'], color='blue', label='Original VWAP')
plt.title('Original VWAP')
plt.xlabel('Time')
plt.ylabel('VWAP')
plt.legend()

# Plot the Box-Cox transformed VWAP
plt.subplot(1, 2, 2)
plt.plot(df.index, df['vwap_boxcox'], color='green', label='Box-Cox Transformed VWAP')
plt.title('Box-Cox Transformed VWAP')
plt.xlabel('Time')
plt.ylabel('Transformed VWAP')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
columns_to_transform = ['Low', 'High', 'Volume', 'Turnover', 'Trades']

lambdas = {}

for col in columns_to_transform:
    
    df[col] = df[col].apply(lambda x: x if x > 0 else 1)  
    
    # Apply Box-Cox transformation
    df[f'{col}_boxcox'], lambdas[col] = stats.boxcox(df[col])
    
    # Print lambda value for the current column
    print(f"Box-Cox lambda for {col}: {lambdas[col]}")
```

```python
df
```

```python
from statsmodels.tsa.stattools import adfuller

# List of columns to check for stationarity
columns_to_check = ['vwap_boxcox','Low_boxcox', 'High_boxcox', 'Volume_boxcox', 'Turnover_boxcox', 'Trades_boxcox']

# Function to perform the Augmented Dickey-Fuller test
def check_stationarity(series, column_name):
    result = adfuller(series.dropna())  # Dropping NA values to ensure test runs correctly
    adf_statistic = result[0]
    p_value = result[1]
    critical_values = result[4]
    print(f"ADF Test for {column_name}:")
    print(f"ADF Statistic: {adf_statistic}")
    print(f"p-value: {p_value}")
    print(f"Critical Values: {critical_values}")
    if p_value < 0.05:
        print(f"The series '{column_name}' is stationary.\n")
    else:
        print(f"The series '{column_name}' is NOT stationary.\n")

# Loop through each column and check for stationarity
for col in columns_to_check:
    check_stationarity(df[col], col)
```

```python
# Define the columns to be differenced
columns_to_difference = ['vwap_boxcox', 'Low_boxcox', 'High_boxcox', 'Turnover_boxcox']

# Loop over each column and apply differencing
for col in columns_to_difference:
    df[f'{col}_diff'] = df[col].diff(30)  # Creates a new column with differenced values

df.dropna(inplace=True)
```

# Feature Engineering

```python
df_original=df_cleaned.copy()
```

## Lag Features

```python
# List of features to create lags for
lag_features = ['Low_boxcox_diff','High_boxcox_diff','Volume_boxcox_diff','Turnover_boxcox_diff','Trades_boxcox_diff']

# Create multiple lags (3,7,30 days) for each feature
for feature in lag_features:
    for lag in [3, 7, 30]:
        df_original[f'{feature}_lag_{lag}'] = df_original[feature].shift(lag)
```

```python
df_original
```

## Moving Averages

```python
for feature in lag_features:
    df_original[f'{feature}_MA_3'] = df_original[feature].rolling(window=3).mean()
    df_original[f'{feature}_MA_7'] = df_original[feature].rolling(window=7).mean()
    df_original[f'{feature}_MA_30'] = df_original[feature].rolling(window=30).mean()
```

```python
df_original.head(10)
```

## Volatility Feature

```python
for feature in lag_features:
    df_original[f'{feature}_volatility_3'] = df_original[feature].rolling(window=3).std()
    df_original[f'{feature}_volatility_7'] = df_original[feature].rolling(window=7).std()
    df_original[f'{feature}_volatility_30'] = df_original[feature].rolling(window=30).std()
```

```python
df_original.head(10)
```

## Time Features

# Modeling

## Resampling

```python
df_original=df_original.resample('MS').sum()
```

## Train-Test Split

```python

train = df_original[df_original.index < "2019"]
valid = df_original[df_original.index >= "2019"]
```

### Resampling done to get rid of noise and be able to visualize trends on higher level

```python
df_original.columns
```

```python
features=[ 'Low_boxcox_diff', 'High_boxcox_diff',
       'Volume_boxcox_diff', 'Turnover_boxcox_diff', 'Trades_boxcox_diff',
       'Low_boxcox_diff_lag_3', 'Low_boxcox_diff_lag_7',
       'Low_boxcox_diff_lag_30', 'High_boxcox_diff_lag_3',
       'High_boxcox_diff_lag_7', 'High_boxcox_diff_lag_30',
       'Volume_boxcox_diff_lag_3', 'Volume_boxcox_diff_lag_7',
       'Volume_boxcox_diff_lag_30', 'Turnover_boxcox_diff_lag_3',
       'Turnover_boxcox_diff_lag_7', 'Turnover_boxcox_diff_lag_30',
       'Trades_boxcox_diff_lag_3', 'Trades_boxcox_diff_lag_7',
       'Trades_boxcox_diff_lag_30', 'Low_boxcox_diff_MA_3',
       'Low_boxcox_diff_MA_7', 'Low_boxcox_diff_MA_30',
       'High_boxcox_diff_MA_3', 'High_boxcox_diff_MA_7',
       'High_boxcox_diff_MA_30', 'Volume_boxcox_diff_MA_3',
       'Volume_boxcox_diff_MA_7', 'Volume_boxcox_diff_MA_30',
       'Turnover_boxcox_diff_MA_3', 'Turnover_boxcox_diff_MA_7',
       'Turnover_boxcox_diff_MA_30', 'Trades_boxcox_diff_MA_3',
       'Trades_boxcox_diff_MA_7', 'Trades_boxcox_diff_MA_30',
       'Low_boxcox_diff_volatility_3', 'Low_boxcox_diff_volatility_7',
       'Low_boxcox_diff_volatility_30', 'High_boxcox_diff_volatility_3',
       'High_boxcox_diff_volatility_7', 'High_boxcox_diff_volatility_30',
       'Volume_boxcox_diff_volatility_3', 'Volume_boxcox_diff_volatility_7',
       'Volume_boxcox_diff_volatility_30', 'Turnover_boxcox_diff_volatility_3',
       'Turnover_boxcox_diff_volatility_7',
       'Turnover_boxcox_diff_volatility_30', 'Trades_boxcox_diff_volatility_3',
       'Trades_boxcox_diff_volatility_7', 'Trades_boxcox_diff_volatility_30',
       'month', 'day']
```

```python
X_train = train[features]
y_train = train['vwap_boxcox_diff']

X_valid = valid[features]
y_valid = valid['vwap_boxcox_diff']
```

```python
print(f'Training set size: {len(X_train)}')
print(f'Validation set size: {len(X_valid)}')
```

## My Awesome, Beautiful, Gorgeous SARIMA Model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
seasonal_order = (1, 1, 1, 12)
model = SARIMAX(y_train, 
                exog=X_train, 
                order=(1, 1, 1), 
                seasonal_order=seasonal_order)
results = model.fit()

# Summarize the model
print(results.summary())
```

# Results and Visualizations

## Final BAAAAM

```python
predictions = results.get_forecast(steps=len(y_valid), exog=X_valid)
predicted_mean = predictions.predicted_mean
predicted_conf_int = predictions.conf_int()

# Evaluate the model
mse = mean_squared_error(y_valid, predicted_mean)
print("Mean Squared Error:", mse)

# Optional: Print forecasted values
print("Forecasted Values:", predicted_mean)

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(y_valid, label='Actual VWAP', color='blue', marker='o', linestyle='-')
plt.plot(predicted_mean, label='Predicted VWAP', color='red', marker='x', linestyle='--')
plt.fill_between(predicted_conf_int.index, predicted_conf_int.iloc[:, 0], predicted_conf_int.iloc[:, 1], color='pink', alpha=0.3)

plt.title('Actual vs Predicted VWAP')

plt.xlabel('Time')
plt.ylabel('VWAP')
plt.legend()
plt.grid()
plt.show()
```
